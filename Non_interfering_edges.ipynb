```python
import networkx as nx
import matplotlib.pyplot as plt
while True:
  G = nx.erdos_renyi_graph(10, 0.15)
  if nx.is_connected(G):
    break
pos = nx.spring_layout(G)
print(pos)
nx.draw(G, pos, with_labels=True)
plt.show()
```

```python
color_dict = {
    0: 'r',    # Red
    1: 'b',    # Blue
    2: 'g',    # Green
    3: 'c',    # Cyan
    4: 'm',    # Magenta
    5: 'y',    # Yellow
    6: 'k',    # Black
    7: 'orange',
    8: 'purple',
    9: 'brown',
    10: 'pink'
}
```

```python
F = nx.line_graph(G)
pos2 = nx.spring_layout(F)
nx.draw(F, pos2, with_labels=True)
plt.show()
```

```python
edges_to_colors = nx.coloring.greedy_color(F, strategy="largest_first")
color_map_nodes = []
color_map_edges = []
for node in F:
  color_map_nodes.append(edges_to_colors[node])

for edge in G.edges():
  color_map_edges.append(edges_to_colors[edge])
pos2 = nx.spring_layout(F)
nx.draw(F, pos2, with_labels=True, node_color=color_map_nodes)
plt.show()
print(f"Number of colors needed {max([x for x in color_map_nodes])}")
```

```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, edge_color=color_map_edges)
plt.show()
```
